```python
# ex 12-5

import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

xdata= ['안지영', '홍지수', '황예린']
ydata1= [90,85,88]
ydata2= [83,88,91]
ydata3= [85,97,78]
ydata4= [92,88,82]

plt.plot(xdata, ydata1,label='국어',color='red',linestyle='-',marker='o')
plt.plot(xdata, ydata2,label='영어',color='#00ffff',linestyle='--',marker='x')
plt.plot(xdata, ydata3,label='수학',color='magenta',linestyle='-.',marker='s')
plt.plot(xdata, ydata4,label='사회',color='#444444',linestyle=':',marker='d')
         
plt.title('세 명 학생의 네 과목 성적')
plt.legend(loc='best')
         
plt.show()
```

```python
plt.plot(xdata,ydata1,'ro-',label='국어', linewidth=3, markersize=10)
plt.plot(xdata,ydata2,'cx--',label='영어', linewidth=3, markersize=10)
plt.plot(xdata,ydata3,'ms-.',label='수학', linewidth=3, markersize=10)
plt.plot(xdata,ydata4,'kd:',label='사회', linewidth=3, markersize=10)
         
plt.title('세 명 학생의 네 과목 성적')
plt.legend(loc='best')
         
plt.show()
```

```python
# ex 12-6

import matplotlib.pyplot as plt

xdata = list(range(30))
ydata = []

for x in xdata:
    y = 2*x
    ydata.append(y)
    
plt.plot(xdata,ydata,'mo-',label='y=2x')
plt.xlim(0,35)
plt.ylim(0,70)
plt.xticks(list(range(0,32,2)))
plt.yticks(list(range(0,65,4)))

plt.grid(True)
plt.show()
```

```python
plt.title('y=2x')
plt.plot(xdata,ydata,'m-', linewidth=3)
plt.xlim(-1,30)
plt.ylim(-2,60)
plt.xticks(list(range(0,30,2)))
plt.yticks(list(range(0,60,5)))

plt.grid(True)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot, rcParams
from math import sin,cos,pi

plt.rcParams['figure.figsize'] = (12,6)


x = np.linspace(0,(2*np.pi), 500)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
x = [0.01*i for i in range(-200,201)]
y = [cos(j*pi)for j in x]

plt.plot(x,y)
plt.show()
```

```python
plt.title = 'cos 삼각함수'

x = np.linspace(0,(2*np.pi), 500)
y = np.cos(x)

plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,(2*np.pi), 500)
y = np.tan(x)

plt.plot(x,y)
plt.show()
```

```python
#Sine, cosine 그래프

import numpy as np

xs = np.arange(0,2*np.pi*0.01,0.01)
ysin = np.sin(xs)
ycos = np.cos(xs)

plt.title('삼각함수 그래프')

plt.plot(xs, ysin, label = 'Sine curve')
plt.plot(xs, ycos, label = 'Cosine curve')

plt.legend()
plt.xticks(np.arange(0, 2 * np.pi * 0.01, np.pi/2))

plt.axhline(y=0, color = 'k', linewidth = 0.5)
plt.axhline(y=1, color = 'k', linewidth = 0.5)
plt.axhline(y=-1, color = 'k', linewidth = 0.5)
plt.axvline(x=0, color = 'k', linewidth = 0.5)
plt.axvline(x=np.pi, color = 'k', linewidth = 0.5)
plt.axvline(x=2*np.pi, color = 'k', linewidth = 0.5)

plt.show()
```

```python
# Sigmoid 함수

xs = np.linspace(-3,3,601)
ys = 1/(1 + np.exp(-xs))    #sigmoid 함수
plt.plot(xs, ys)
plt.axhline(y=0, color = 'k', linewidth = 0.5)
plt.axhline(y=1, color = 'k', linewidth = 0.5)
plt.axvline(x=0, color = 'k', linewidth = 0.5)
plt.show()

```
