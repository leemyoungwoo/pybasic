### 데이터 조작

```python
# ex 14-11

scores = {'이름':['김지영','안지수','최성수','황예린','김소정'],
         '국어':[95,97,90,94,87],
         '영어':[90,86,93,85,93],
         '수학':[85,88,89,88,99]}
```

```python
df = pd.DataFrame(scores)

df
```

```python
df2 = df.iloc[:,[1,2,3]]
df2
```

```python
df['합계'] = df2.sum(axis=1) # 합계를 '열'에 추가하여 의미있게 사용

df
```

```python
total2 = df2.sum(axis=0)

total2
```

```python
# ex 14-12 

scores = {'이름':['김지영','안지수','최성수','황예린','김소정'],
         '국어':[95,97,90,94,87],
         '영어':[90,86,93,85,93],
         '수학':[85,88,89,88,99]}
```

```python
df = pd.DataFrame(scores)

df
```

```python
df2 = df.iloc[:,[1,2,3]]

df2
```

```python
df['합계']= df2.sum(axis=1)

df
```

```python
df['평균'] = df2.mean(axis=1)

df
```

```python
df[['이름','합계','평균']]
```
