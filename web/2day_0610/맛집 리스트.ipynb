```python
url = "https://www.tripadvisor.co.kr/RestaurantsNear-g2024867-d9579419-Daerim-Gwangmyeong_Gyeonggi_do.html"
html = urlopen(url)
soup = BeautifulSoup(html,'html.parser')
```

```python
rest_list = soup.find_all(class_='near_listing')
rest_list
```

```python
len(rest_list)
```

```python
tmp = rest_list[0]
tmp
```

```python
tmp.find_all('a')[1].get_text()
```

```python
tmp.find_all(class_='format_address')[0].get_text()
```

```python
name = []
addr = []
for rest in rest_list:
    name.append(rest.find_all('a')[1].get_text())
    addr.append(rest.find_all(class_='format_address')[0].get_text())
```

```python
rest_list = pd.DataFrame({'상호명':name, '주소':addr})
rest_list
```
